# 字典学习
* 字典就是 映射 ，键值对。相当于java中的map

```python

# 这是一个字典
people={
      'Alice':{
      'phone':'2341',
      'addr':'KA'
      },
      'Beth':{
      'phone':'9102',
      'addr':'USA'
      },
      'Cecil':{
      'phone':'3158',
      'addr':'UK'
      }
}
#电话号码
labels={
   'phone':'phone number',
   'addr':'address'
}
name=input('Name: ')
# 要查找电话号码号码还是地址
request=input('Phone number (p) or address (a)? ')
# 使用正确的键

if request=='p':key='phone'
if request=='a':key='addr'

if name in people:
    print("%s's %s is %s." % (name, labels[key], people[name][key]))
```

## 字典的基本操作
* len(d)返回字典d包含的项（键-值对） 数
* d[k]返回与键k相关联的值
* d[k]=v 将值v关联为k的项
* del d[k] 删除键为k的项
* k in d 检查字典d是否包含键为k的项
* clear
* copy
* fromkeys
* get

```python
# 一个使用 get()的简单数据库
people={
      'Alice':{
      'phone':'2341',
      'addr':'KA'
      },
      'Beth':{
      'phone':'9102',
      'addr':'USA'
      },
      'Cecil':{
      'phone':'3158',
      'addr':'UK'
      }
}
#电话号码
labels={
   'phone':'phone number',
   'addr':'address'
}
name=input('Name: ')

#要查找电话号码还是地址
request= input('Phone number (p) or address (a)? ')

#使用正确的键
key= request 
if request=='p' : key ='phone'
if request=='a' : key='addr'
# 使用 get 提供默认值
person=people.get(name,{})
label =labels.get(key,key)
result=person.get(key,'not available')

print("{}'s {} is {}".format(name,label,result) )
```

## 安全访问字典键
避免 KeyError，可先尝试访问并提供默认值。下面的例子同时展示 try/except 与 get()/setdefault() 的对比。

```python
phone_book = {'alice': '10086', 'bob': '10010'}
try:
    print('Bob 的号码:', phone_book['bob'])
    print('Tom 的号码:', phone_book['tom'])
except KeyError:
    print('未找到 tom，改用 get:', phone_book.get('tom', '未知号码'))
```

```python
contacts = {}
print('初始字典:', contacts)
# setdefault 会在键不存在时写入默认值
contacts.setdefault('alice', []).append('10086')
contacts.setdefault('alice', []).append('66666')
print('使用 setdefault 后:', contacts)
```

### 练习：电话号码查找
- 将名字转换为小写后再查询，避免大小写差异。
- 使用 get() 提供友好的提示信息。
- 可扩展：添加联系人时，使用 setdefault 维护一个号码列表。

```python
phone_book = {'alice': '10086', 'bob': '10010'}

def lookup(name: str) -> None:
    number = phone_book.get(name.lower())
    if number:
        print(f'{name} 的号码是 {number}')
    else:
        print(f'未找到 {name}，试着新增一条记录吧！')

lookup('Alice')
lookup('Tom')
```
