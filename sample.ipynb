# 字典学习
* 字典就是 映射 ，键值对。相当于java中的map

```python

# 这是一个字典
people={
      'Alice':{
      'phone':'2341',
      'addr':'KA'
      },
      'Beth':{
      'phone':'9102',
      'addr':'USA'
      },
      'Cecil':{
      'phone':'3158',
      'addr':'UK'
      }
}
#电话号码
labels={
   'phone':'phone number',
   'addr':'address'
}
name=input('Name: ')
# 要查找电话号码号码还是地址
request=input('Phone number (p) or address (a)? ')
# 使用正确的键

if request=='p':key='phone'
if request=='a':key='addr'

if name in people:
    print("%s's %s is %s." % (name, labels[key], people[name][key]))
```

## 字典的基本操作
* len(d)返回字典d包含的项（键-值对） 数
* d[k]返回与键k相关联的值
* d[k]=v 将值v关联为k的项
* del d[k] 删除键为k的项
* k in d 检查字典d是否包含键为k的项
* clear
* copy
* fromkeys
* get

```python
# 一个使用 get()的简单数据库
people={
      'Alice':{
      'phone':'2341',
      'addr':'KA'
      },
      'Beth':{
      'phone':'9102',
      'addr':'USA'
      },
      'Cecil':{
      'phone':'3158',
      'addr':'UK'
      }
}
#电话号码
labels={
   'phone':'phone number',
   'addr':'address'
}
name=input('Name: ')

#要查找电话号码还是地址
request= input('Phone number (p) or address (a)? ')

#使用正确的键
key= request 
if request=='p' : key ='phone'
if request=='a' : key='addr'
# 使用 get 提供默认值
person=people.get(name,{})
label =labels.get(key,key)
result=person.get(key,'not available')

print("{}'s {} is {}".format(name,label,result) )
```
